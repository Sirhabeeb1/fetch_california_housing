CREATING THE INDEPENDENT AND DEPENDENT VARIABLE 

```python
x = df1.iloc[:,:-1]
```

```python
y = df1["Target"]
```

SPLITTING THE DATASET USING SKLEARN

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

STANDARDIZING THE X_TRAIN AND X_TEXT USING STANDARDSCALER METHOD

```python

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# x_train, y_train, x_test, y_test are already defined
pipeline = make_pipeline(StandardScaler(), LinearRegression())

# Training the model
model = pipeline.fit(x_train, y_train)
```

Making predictions using x_test

```python
y_predict = model.predict(x_test) 
```

Evaluating the model

```python
mse = mean_squared_error(y_test, y_predict)  # Corrected from "y_text"
print(f"Mean Squared Error: {mse}")

mae = mean_absolute_error(y_test, y_predict)
print(f'Mean Absolute Error: {mae}')

r2 = r2_score(y_test, y_predict)
print(f"R² Score: {r2}")
```

Displaying model parameters

```python
display(model.get_params())
```

Scatter plot of actual vs predicted values

```python
plt.scatter(y_test, y_predict)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.show()
```

Calculating the residuals and  Residual plot using (KDE distribution)

```python
residuals = y_test - y_predict

sns.displot(residuals, kind="kde")
plt.title("Residual Distribution")
plt.show()
```

Reshaping `df_pred` correctly before making predictions

```python
pred = x.iloc[1, :].values.reshape(1,-1)
df1_pred_scaled = model.predict(pred)
print("Predicted value:", df1_pred_scaled)
```
